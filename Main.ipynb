```julia
Δ_SC = 1.0
Δ_Z = 1.1*Δ_SC

θ = pi/4

Δ_I(β) = Δ_SC - Δ_Z*sin(β)
Δ_II(β) = Δ_SC - Δ_Z*cos(β)
Δ_III(β) = Δ_SC - Δ_Z*cos(abs(β - θ))
```

```julia
MZM_at_corner1(β) =  (Δ_I(β) > 0 && Δ_II(β) < 0) || (Δ_I(β) < 0 && Δ_II(β) > 0) ? true : false
MZM_at_corner2(β) =  (Δ_III(β) > 0 && Δ_I(β) < 0) || (Δ_III(β) < 0 && Δ_I(β) > 0) ? true : false
MZM_at_corner3(β) =  (Δ_II(β) > 0 && Δ_III(β) < 0) || (Δ_II(β) < 0 && Δ_III(β) > 0) ? true : false
```

```julia
β = 0

@show Δ_I(β) 
@show Δ_II(β) 
@show Δ_III(β)

@show MZM_at_corner1(β)
@show MZM_at_corner2(β)
@show MZM_at_corner3(β)
```

```julia
β_values = LinRange(0, pi/2, 100)

for β = β_values
    @show Δ_I(β) 
    @show Δ_II(β) 
    @show Δ_III(β)
    
    @show MZM_at_corner1(β)
    @show MZM_at_corner2(β)
    @show MZM_at_corner3(β)
end
```
